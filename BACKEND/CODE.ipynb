```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Target'] = le.fit_transform(df['Target'])
```

```python
df = df.drop('Marital status', axis=1)
```

```python
x = df.drop('Target', axis=1) 
y = df['Target']
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
x,y = smote.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_selection import SelectKBest, f_classif 

selected = SelectKBest(score_func=f_classif, k = 15)
x_selected = selected.fit_transform(x,y)
```

```python
selected_index = selected.get_support(indices=True)
print(selected_index)
```

```python
selected_columns = x.columns[selected.get_support()] 
print(selected_columns.to_list())
```

```python
df_selected = pd.DataFrame(x_selected, columns=selected_columns)
df_selected['Class'] = y.values
```

```python
df_selected.to_csv('DATASET/final_dataset.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_selected,y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neural_network import MLPClassifier 
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
mlp = MLPClassifier()
mlp.fit(x_train, y_train)
```

```python
pred = mlp.predict(x_train)
training_accuracy = accuracy_score(y_train, pred)
print("Training accuracy of MLP : ", training_accuracy)
report = classification_report(y_train, pred) 
print("classification_report:\n", report)
conf_matrix = confusion_matrix(y_train, pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
y_pred_test = mlp.predict(x_test) 
test_accuracy = accuracy_score(y_test, y_pred_test)
print("Testing accuracy score of mlp : ", test_accuracy)
report = classification_report(y_test, y_pred_test) 
print("classification_report:\n", report)
conf_matrix = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Import necessary libraries
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Define base models
base_model = [
    ('rf', RandomForestClassifier()),
    ('dt', DecisionTreeClassifier())
]

# Define meta-model
meta_model = LogisticRegression()

# Define Stacking Classifier
stc = StackingClassifier(estimators=base_model, final_estimator=meta_model)

# Hyperparameter grids for tuning
param_grid = {
    'rf__n_estimators': [100, 200, 300],
    'rf__max_depth': [None, 10, 20],
    'dt__max_depth': [None, 10, 20],
    'dt__min_samples_split': [2, 5, 10],
    'final_estimator__C': [0.1, 1, 10],
    'final_estimator__solver': ['liblinear', 'saga']
}

# Use GridSearchCV to tune hyperparameters
grid_search = GridSearchCV(estimator=stc, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1, scoring='accuracy')

# Fit the grid search to the training data
grid_search.fit(x_train, y_train)

# Best parameters from the grid search
print("Best parameters from GridSearchCV: ", grid_search.best_params_)

# Get the best estimator
best_model = grid_search.best_estimator_

# Predict on the test set using the best model
y_pred_test = best_model.predict(x_test)

# Evaluate the model performance
test_accuracy = accuracy_score(y_test, y_pred_test)
print("Testing accuracy score of Stacking Classifier after hyperparameter tuning: ", test_accuracy)

# Classification report
report = classification_report(y_test, y_pred_test)
print("Classification Report:\n", report)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initialize base classifiers
log_clf = LogisticRegression(max_iter=200)
rf_clf = RandomForestClassifier(n_estimators=100)
knn_clf = KNeighborsClassifier(n_neighbors=3)

# Create a voting classifier (VotingClassifier)
vtc = VotingClassifier(estimators=[('lr', log_clf), ('rf', rf_clf), ('knn', knn_clf)],
                              voting='hard') 
vtc.fit(x_train, y_train)
```

```python
pred = vtc.predict(x_train)
training_accuracy = accuracy_score(y_train, pred)
print("Training accuracy of Voting Classifier : ", training_accuracy)
report = classification_report(y_train, pred) 
print("classification_report:\n", report)
conf_matrix = confusion_matrix(y_train, pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
y_pred_test = vtc.predict(x_test) 
test_accuracy = accuracy_score(y_test, y_pred_test)
print("Testing accuracy score of Voting Classifier : ", test_accuracy)
report = classification_report(y_test, y_pred_test) 
print("classification_report:\n", report)
conf_matrix = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
